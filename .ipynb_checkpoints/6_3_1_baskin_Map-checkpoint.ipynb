```python
# 예제 6-51 라이브러리 임포트
import pandas as pd
import folium
import json
```

```python
# 예제 6-53 서울시 스타벅스 매장 목록 데이터 불러오기
seoul_baskin = pd.read_excel(filepath)
seoul_baskin.tail()
```

```python
# 예제 6-54 folium을 이용한 지도 생성
baskin_map = folium.Map(
    location=[37.573050, 126.979189],
    tiles='Stamen Terrain',
    zoom_start=11
)
baskin_map
```

```python
# 예제 6-55 지도에 스타벅스 매장 위치를 나타내는 서클 마커 그리기
for idx in seoul_baskin.index:
    lat = seoul_baskin.loc[idx, '위도']
    lng = seoul_baskin.loc[idx, '경도']

    folium.CircleMarker(
        location=[lat, lng],
        fill = True, 
        fill_color='green', 
        fill_opacity=1,
        color='yellow', 
        weight=1,
        radius=3
    ).add_to(baskin_map)
baskin_map
```

```python
# 예제 6-56 스타벅스 매장 타입별 위치 서클 마커 그리기
baskin_map2 = folium.Map(
    location=[37.573050, 126.979189],
    tiles='Stamen Terrain',
    zoom_start=11
)

for idx in seoul_baskin.index:
    lat = seoul_baskin.loc[idx, '위도']
    lng = seoul_baskin.loc[idx, '경도']
    store_type = seoul_baskin.loc[idx, '별점']
    type(store_type)

    # 매장 타입별 색상 선택을 위한 조건문
    fillColor = ''
    if store_type < 3:
        fillColor = 'black'
        size = 2
    elif 3 <= store_type <= 4:
        fillColor = 'blue'
        size = 5
    elif store_type > 4:
        fillColor = 'red'
        size = 5

    folium.CircleMarker(
        location=[lat, lng],
        color=fillColor,
        fill = True,
        fill_color = fillColor, 
        fill_opacity = 1,
        weight = 1,
        radius = size
    ).add_to(baskin_map2)

baskin_map2
```
