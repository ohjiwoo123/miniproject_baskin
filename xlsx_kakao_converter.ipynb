```python
import requests
import json
import pandas as pd
```

병합할 엑셀파일 읽어오기

```python
# 잘 읽어왔는지 체크
df = pd.read_excel(inpath)  
```

```python
# 특정 열(칼럼) 읽어와보기
df_sample = df.iloc[:,[2]]
df_sample
```

```python
# 주소 전달
print(df.iloc[0]['주소'])
```

```python
len(df_sample)
```

카카오 api


```python
# 카카오 내 키 
headers = {"Authorization": "KakaoAK APIKEY"}
```

```python
# 좌표값 저장할 리스트 생성
lat = []
lng = []
count = 0

for idx in range(0,len(df_sample)):
  address = df.iloc[idx]['주소']
  url = "https://dapi.kakao.com/v2/local/search/address.json?&query=" + address

  api_test = requests.get(url,headers=headers)
  url_text = json.loads(api_test.text)

  count += 1
  print(count)
  lat.append (url_text.get('documents')[0].get('y'))
  lng.append (url_text.get('documents')[0].get('x'))

lat, lng
```

```python
df[180:190]
```

```python
# 리스트를 이용해 merge할 좌표 데이터프레임 생성 
merge_df = pd.DataFrame(data = list(zip(lat, lng)), columns = ['위도','경도'])
```

```python
merge_df
```

```python
# 원래 데이터프레임 df와 좌표열이 있는 merge_df를 병합하기
result_df = pd.concat([df, merge_df],axis=1)
```

```python
result_df
```

```python
## 데이터 저장 /files/seoul_beskinrobins_list2.xlsx
result_df.to_excel(outpath, index=False)
```
